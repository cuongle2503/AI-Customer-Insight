```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_path = 'sales_06_FY2020-21.csv'
data = pd.read_csv(file_path, dtype={'order_id': str})


data_info = data.info()
```

```python
data_head = data.head(10)
data_head.T
```

```python
data_cleaned = data.copy()
data_cleaned.shape
```

```python
data_cleaned['qty_ordered'] = data_cleaned['qty_ordered'].astype(str)
data_cleaned['price'] = data_cleaned['price'].astype(str)
data_cleaned['total'] = data_cleaned['total'].astype(str)
```

### Xử lý tiền tệ và dữ liệu dạng số

```python
data_cleaned['qty_ordered'] = pd.to_numeric(data_cleaned['qty_ordered'], errors='coerce')

data_cleaned['price'] = pd.to_numeric(data_cleaned['price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
data_cleaned['total'] = pd.to_numeric(data_cleaned['total'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
```

### Xử lý dữ liệu trống

```python
data_cleaned.dropna(subset=['qty_ordered'], inplace=True)  

data_cleaned.dropna(subset=['price'], inplace=True)

data_cleaned['value'] = data_cleaned['value'].fillna(data_cleaned['qty_ordered'] * data_cleaned['price'])

data_cleaned['total'] = np.where(
    data_cleaned['discount_amount'] != 0,  # Điều kiện
    data_cleaned['discount_amount'] * data_cleaned['value'],  # Nếu điều kiện đúng
    data_cleaned['value']  # Nếu điều kiện sai
)
```

### Xoá dữ liệu bị trùng

```python
data_cleaned = data_cleaned.drop_duplicates(subset=data_cleaned.columns.difference(['order_id']))
```
